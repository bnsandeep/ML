#### Matplotlib: Visualization with Python
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

### Basic Matplotlib commands

```python
plt.plot(x,y,'r'); #'r' specifies red color

plt.xlabel('X Axis title here')
plt.ylabel('Y Axis title here')
plt.title('Plot Title')
plt.show()
```

### Create Multiplots on same canvas

```python
plt.subplot(1,4,1)  # nrows, ncols, plot_number
plt.plot(x,y,'r')
plt.subplot(1,4,2)  # nrows, ncols, plot_number
plt.plot(y,x,'g');
plt.subplot(1,4,3)  # nrows, ncols, plot_number
plt.plot(x,x,'b');
plt.subplot(1,4,4)  # nrows, ncols, plot_number
plt.plot(y,y**3,'b');
```

### Matplotlib object oriented method

```python
### Create figure

fig = plt.figure()

# Add axes

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height

axes.plot(x,y,'b')
axes.set_xlabel('X Label Title')
axes.set_ylabel('Y Label Title')
axes.set_title('Plot Title');
```

```python
fig = plt.figure(figsize=(12,6))

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height

axes1.plot(x,y,'b')
axes1.set_xlabel('X Label Title')
axes1.set_ylabel('Y Label Title')
axes1.set_title('Plot Title');

axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # left, bottom, width, height

axes2.plot(y,x,'r')
axes2.set_xlabel('X Label Title')
axes2.set_ylabel('Y Label Title')
axes2.set_title('Plot Title');
```

### Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")

ax.legend(loc=4)
```

#### Setting colors, linewidths, linetypes

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, 'b.-')
ax.plot(x, x**3, 'g--');
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color='blue',alpha=0.5) # alpha determines transparency
ax.plot(x, x+2, color='#8B008B');
ax.plot(x, x+3, color='#784a96');
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color='blue',alpha=0.5,linewidth=1.25) # alpha determines transparency
ax.plot(x, x+2, color='#8B008B',linewidth=3.5);
ax.plot(x, x+3, color='#784a96',linewidth=3);
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+4, color='blue',alpha=0.5,linewidth=5.25,linestyle='-') # alpha determines transparency
ax.plot(x, x+1, color='blue',linewidth=5.25,linestyle='-') # alpha determines transparency
ax.plot(x, x+2, color='#8B008B',linewidth=3.5,linestyle='-.');
ax.plot(x, x+3, color='#784a96',linewidth=3,linestyle=':');
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+4, color='blue',alpha=0.5,linewidth=5.25,linestyle='-',marker='o',markersize=3) # alpha determines transparency
ax.plot(x, x+1, color='blue',linewidth=5.25,linestyle='-',marker='x',markersize=5) # alpha determines transparency
ax.plot(x, x+2, color='#8B008B',linewidth=3.5,linestyle='-.',marker='s',markersize=6);
ax.plot(x, x+3, color='#784a96',linewidth=3,linestyle=':',marker='1',markersize=7);
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+4, color='red',alpha=0.5,
        linewidth=5.25,
        linestyle='-',
        marker='o',
        markersize=8,
        markerfacecolor='yellow',
        markeredgecolor='green',
        markeredgewidth=4)
```
